# Pred - Prey Model Functions

```python
def dNdt(H, P, parameters_dict):
    #Unpacking herbivore parameters:
    r_H = parameters_dict['r_H'] # growth rate of herbivores
    K_H = parameters_dict['K_H'] # carrying capacity of herbivores
    theta_H = parameters_dict['theta_H'] # immigration of herbivores
    c = parameters_dict['c'] # addition to predator from prey
    d = parameters_dict['d'] # value of H when predation half its max value
    q_H = parameters_dict['q_H'] # proportionality constant for herbivore harvest
    E_H = parameters_dict['E_H'] # Effort for herbivore harvest
    #Unpacking predator parameters:
    r_P = parameters_dict['r_P'] # growth rate of predator 
    K_P = parameters_dict['K_P'] # carrying capactiy of predators 
    b = parameters_dict['b'] # mortality to prey from predator
    theta_P = parameters_dict['theta_P'] # immigration of predators  
    q_P = parameters_dict['q_P'] # proportionality constant for predators harvest
    E_P = parameters_dict['E_P'] # Effort for predator harvest
    
    #Calculate the derivative 
    dH = r_H*H*(1-H/K_H) - (c*H*P)/(d+H) + theta_H - q_H*E_H*H
    dP = r_P*P*(1-P/K_P) + (b*H*P)/(d+H) + theta_P - q_P*E_P*P
    
    return dH, dP
```

```python
# Function: numerical solver, 2nd order RK 

def RK2(H, P, parameters_dict):
    H_init = H
    P_init = P 
    
    dH1, dP1 = dNdt(H,P,parameters_dict)
    
    H1 = H + 0.5*dH1
    P1 = P + 0.5*dP1
    
    dH2, dP2 = dNdt(H1,P1,parameters_dict)
    
    dH_ave = (dH1+dH2)/2 
    dP_ave = (dP1+dP2)/2 
    
    H = H_init + dH_ave
    P = P_init + dP_ave
    
    #Calculate harvest 
    q_H = parameters_dict['q_H'] #proportionality constant for herbivore harvest 
    E_H = parameters_dict['E_H'] #Effort for herbivore harvest
    
    q_P = parameters_dict['q_P'] #proportionality constant for herbivore harvest 
    E_P = parameters_dict['E_P'] #Effort for herbivore harvest
    
    herbivore_harvest = q_H*E_H*H
    predator_harvest = q_P*E_P*P
    
    return H, P, herbivore_harvest, predator_harvest
```

```python
# Function to run the model 

def run_model(INIT_H, INIT_P, parameters_dict):
    
    numyears = parameters_dict['numyears']
    
    H_array = np.zeros(numyears+1)
    P_array = np.zeros(numyears+1)
    H_harvest_array = np.zeros(numyears+1)
    P_harvest_array = np.zeros(numyears+1)
    
    H_array[0] = INIT_H #Starting biomass of herbivore 
    P_array[0] = INIT_P #Starting biomass of predator 
    H_harvest_array[0] = 0
    P_harvest_array[0] = 0
    
    for i in np.arange(0,numyears):
        H_array[i+1], P_array[i+1], H_harvest_array[i+1], P_harvest_array[i+1] \
            = RK2(H_array[i], P_array[i], parameters_dict)
        i = i+1
    
    return H_array, P_array, H_harvest_array, P_harvest_array
```

# Quantifying Limit Cycles

```python
def get_cycle_characteristics(H_array, P_array, numyears, transient=50):
    from scipy.signal import find_peaks

    # Remove transient dynamics
    H_trim = H_array[transient:]
    P_trim = P_array[transient:]
    
    # Find peaks and compute amplitude
    H_peaks, _ = find_peaks(H_trim)
    P_peaks, _ = find_peaks(P_trim)
    
    if len(H_peaks) < 2 or len(P_peaks) < 2:
        return None, None  # No clear cycle

    # Amplitude = mean peak - mean trough
    H_amp = np.mean(H_trim[H_peaks]) - np.min(H_trim)
    P_amp = np.mean(P_trim[P_peaks]) - np.min(P_trim)
    
    # Period = average time between peaks
    H_period = np.mean(np.diff(H_peaks))
    P_period = np.mean(np.diff(P_peaks))
    
    return (H_amp, H_period), (P_amp, P_period)
```

```python
b = c = 0.6
E_P_vals = np.linspace(0, 1, 30)
H_amps, H_periods = [], []

for ep in E_P_vals:
    parameters_dict['b'] = b
    parameters_dict['c'] = c
    parameters_dict['E_P'] = ep
    
    H_array, P_array, _, _ = run_model(H0, P0, parameters_dict)
    H_metrics, _ = get_cycle_characteristics(H_array, P_array, numyears)
    
    if H_metrics is not None:
        amp, period = H_metrics
        H_amps.append(amp)
        H_periods.append(period)
    else:
        H_amps.append(0)
        H_periods.append(0)

# Plot amplitude and period
fig, ax = plt.subplots(2, 1, figsize=(8, 6), sharex=True)
ax[0].plot(E_P_vals, H_amps, marker='o')
ax[0].set_ylabel('Prey Amplitude')

ax[1].plot(E_P_vals, H_periods, marker='o')
ax[1].set_ylabel('Prey Period')
ax[1].set_xlabel('Predator Harvest (E_P)')

plt.title('Limit Cycle Characteristics Across Predator Harvest Levels')
plt.tight_layout()
plt.show()
```

```python
def get_cycle_characteristics(H_array, P_array, numyears, transient=50):
    from scipy.signal import find_peaks

    # Remove transient
    H_trim = H_array[transient:]
    P_trim = P_array[transient:]

    # Prey metrics
    H_peaks, _ = find_peaks(H_trim)
    H_amp = np.mean(H_trim[H_peaks]) - np.min(H_trim) if len(H_peaks) >= 2 else 0
    H_period = np.mean(np.diff(H_peaks)) if len(H_peaks) >= 2 else 0

    # Predator metrics
    P_peaks, _ = find_peaks(P_trim)
    P_amp = np.mean(P_trim[P_peaks]) - np.min(P_trim) if len(P_peaks) >= 2 else 0
    P_period = np.mean(np.diff(P_peaks)) if len(P_peaks) >= 2 else 0

    return (H_amp, H_period), (P_amp, P_period)


E_P_vals = np.linspace(0, 1.0, 25)
prey_periods, predator_periods = [], []
prey_amps, predator_amps = [], []

# Set high interaction strength
parameters_dict['b'] = 0.6
parameters_dict['c'] = 0.6
parameters_dict['E_H'] = 0.0

for ep in E_P_vals:
    parameters_dict['E_P'] = ep
    H_array, P_array, _, _ = run_model(H0, P0, parameters_dict)
    H_metrics, P_metrics = get_cycle_characteristics(H_array, P_array, numyears)

    # Unpack and store
    H_amp, H_period = H_metrics
    P_amp, P_period = P_metrics

    prey_amps.append(H_amp)
    prey_periods.append(H_period)
    predator_amps.append(P_amp)
    predator_periods.append(P_period)

    
fig, axs = plt.subplots(2, 1, figsize=(8, 6), sharex=True)

axs[0].plot(E_P_vals, predator_amps, marker='o', color='darkorange')
axs[0].set_ylabel('Predator Amplitude')

axs[1].plot(E_P_vals, predator_periods, marker='o', color='indianred')
axs[1].set_ylabel('Predator Period')
axs[1].set_xlabel('Predator Harvest Intensity (E_P)')

plt.suptitle('Predator Oscillation Dynamics Across Harvest Intensity')
plt.tight_layout()
plt.show()
```

```python
from scipy.signal import find_peaks

# --- Cycle detection ---
def detect_cycles(ts, min_peaks=2, min_amplitude=10, tail_length=100):
    ts = np.asarray(ts)[-tail_length:]
    if (np.max(ts) - np.min(ts)) < min_amplitude:
        return False

    ts_centered = ts - np.mean(ts)
    peaks, _ = find_peaks(ts_centered, height=min_amplitude / 2)
    troughs, _ = find_peaks(-ts_centered, height=min_amplitude / 2)

    return (len(peaks) >= min_peaks) and (len(troughs) >= min_peaks)

# Sweep over E_P and plot asymptotic values with gray amplitude shading
E_P_vals = np.linspace(0, 1.0, 200)
H_traj_points, P_traj_points = [], []
E_P_record = []

# Store min/max values for shading
E_P_cycles = []
H_min_vals, H_max_vals = [], []
P_min_vals, P_max_vals = [], []

# Set fixed high interaction strength
parameters_dict['b'] = 0.6
parameters_dict['c'] = 0.6
parameters_dict['E_H'] = 0.0

transient = 50  # discard first 50 years
sample_window = 50  # use final 50 years of data

for ep in E_P_vals:
    parameters_dict['E_P'] = ep
    H_array, P_array, _, _ = run_model(H0, P0, parameters_dict)

    H_tail = H_array[-sample_window:]
    P_tail = P_array[-sample_window:]

    # Record points for scatter
    H_traj_points.extend(H_tail)
    P_traj_points.extend(P_tail)
    E_P_record.extend([ep] * sample_window)

    # Check for cycles and save amplitude bounds if present
    if detect_cycles(H_tail) or detect_cycles(P_tail):
        E_P_cycles.append(ep)
        H_min_vals.append(np.min(H_tail))
        H_max_vals.append(np.max(H_tail))
        P_min_vals.append(np.min(P_tail))
        P_max_vals.append(np.max(P_tail))

# Plotting
fig, axs = plt.subplots(2, 1, figsize=(8, 6), sharex=True)

# Prey
axs[0].scatter(E_P_record, H_traj_points, s=2, color='green')
axs[0].fill_between(E_P_cycles, H_min_vals, H_max_vals, color='gray', alpha=0.5)
axs[0].set_ylabel('Prey (H) Biomass')
axs[0].set_title('Bifurcation Diagram: Prey & Predator vs. Predator Harvest Intensity')

# Predator
axs[1].scatter(E_P_record, P_traj_points, s=2, color='darkred')
axs[1].fill_between(E_P_cycles, P_min_vals, P_max_vals, color='gray', alpha=0.5)
axs[1].set_ylabel('Predator (P) Biomass')
axs[1].set_xlabel('Predator Harvest Intensity (E_P)')

plt.tight_layout()
plt.show()
```

```python
E_P_vals = np.linspace(0, 1.0, 200)
sample_window = 50
min_cycle_amp = 1.0

# Store full clouds for scatter
P_traj_points = []
E_P_record = []

# For outline curves
P_means = []
P_maxs = []
P_mins = []

parameters_dict['b'] = 0.6
parameters_dict['c'] = 0.6
parameters_dict['E_H'] = 0.0

for ep in E_P_vals:
    parameters_dict['E_P'] = ep
    H_array, P_array, _, _ = run_model(H0, P0, parameters_dict)
    P_tail = P_array[-sample_window:]

    # Record cloud
    P_traj_points.extend(P_tail)
    E_P_record.extend([ep] * sample_window)

    # Envelope detection
    amp = np.ptp(P_tail)
    is_cyclic = detect_cycles(P_tail)

    if is_cyclic and amp >= min_cycle_amp:
        P_means.append(np.nan)  # gap in solid line
        P_maxs.append(np.max(P_tail))
        P_mins.append(np.min(P_tail))
    else:
        P_means.append(np.mean(P_tail))
        P_maxs.append(np.nan)
        P_mins.append(np.nan)

# Convert to arrays
P_means = np.array(P_means)
P_maxs = np.array(P_maxs)
P_mins = np.array(P_mins)

# Plotting
plt.figure(figsize=(8, 5))

# Cloud of values
plt.scatter(E_P_record, P_traj_points, s=2, color='darkred', alpha=0.3, label='Final dynamics')

# Stable mean
plt.plot(E_P_vals, P_means, 'k-', label='Stable equilibrium')

# Min/max dashed lines
plt.plot(E_P_vals, P_maxs, 'k--', linewidth=1.2, label='Cycle bounds')
plt.plot(E_P_vals, P_mins, 'k--', linewidth=1.2)

plt.xlabel('Predator Harvest Intensity (E_P)')
plt.ylabel('Predator Biomass (P*)')
plt.title('Bifurcation Diagram: Predator vs. E_P')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# === Define function to compute bifurcation data ===
def compute_bifurcation_data(param_name, param_vals, fixed_params, H0, P0, sample_window=100, min_cycle_amp=1.0):
    H_means, H_maxs, H_mins, H_cycle_center = [], [], [], []
    P_means, P_maxs, P_mins, P_cycle_center = [], [], [], []

    for val in param_vals:
        params = fixed_params.copy()
        params[param_name] = val

        H_array, P_array, *_ = run_model(H0, P0, params)
        H_tail = H_array[-sample_window:]
        P_tail = P_array[-sample_window:]

        H_amp = np.ptp(H_tail)
        P_amp = np.ptp(P_tail)

        if H_amp < min_cycle_amp:
            H_means.append(np.mean(H_tail))
            H_maxs.append(np.nan)
            H_mins.append(np.nan)
            H_cycle_center.append(np.nan)
        else:
            H_means.append(np.nan)
            H_maxs.append(np.max(H_tail))
            H_mins.append(np.min(H_tail))
            H_cycle_center.append(np.mean(H_tail))

        if P_amp < min_cycle_amp:
            P_means.append(np.mean(P_tail))
            P_maxs.append(np.nan)
            P_mins.append(np.nan)
            P_cycle_center.append(np.nan)
        else:
            P_means.append(np.nan)
            P_maxs.append(np.max(P_tail))
            P_mins.append(np.min(P_tail))
            P_cycle_center.append(np.mean(P_tail))

    return {
        'param_vals': np.array(param_vals),
        'H_means': np.array(H_means),
        'H_maxs': np.array(H_maxs),
        'H_mins': np.array(H_mins),
        'H_cycle_center': np.array(H_cycle_center),
        'P_means': np.array(P_means),
        'P_maxs': np.array(P_maxs),
        'P_mins': np.array(P_mins),
        'P_cycle_center': np.array(P_cycle_center),
    }

# === Parameter sweep setup ===
param_sweeps = {
    #'E_H': np.linspace(0, 1.0, 200),
    'K_H': np.linspace(100, 7000, 200),
    'K_P': np.linspace(100, 7000, 200),
    'b': np.linspace(0, 1.0, 200),
    'c': np.linspace(0, 1.0, 200),
    'r_H': np.linspace(0, 1.0, 200),
    'r_P': np.linspace(0, 1.0, 200),
}

# === Default parameters ===
default_params = {
    'numyears': 400,
    'r_H': 0.4,
    'K_H': 5000,
    'theta_H': 50,
    'c': 0.4,
    'd': 500,
    'q_H': 1.0,
    'E_H': 0.0,
    'r_P': 0.2,
    'K_P': 2000,
    'b': 0.4,
    'theta_P': 10,
    'q_P': 1.0,
    'E_P': 0.0,
}

fig, axs = plt.subplots(3, 2, figsize=(14, 12))
axs = axs.flatten()

for idx, (param_name, param_vals) in enumerate(param_sweeps.items()):
    data = compute_bifurcation_data(param_name, param_vals, default_params, H0=100, P0=10)

    ax = axs[idx]
    
    # Plot prey (N)
    ax.plot(data['param_vals'], data['H_means'], '-', label='N: Stable', color='limegreen')
    ax.plot(data['param_vals'], data['H_cycle_center'], '--', label='N: Cycle Mean', color='limegreen')
    ax.fill_between(data['param_vals'], data['H_mins'], data['H_maxs'],
                    where=~np.isnan(data['H_cycle_center']), color='gray', alpha=0.3)

    # Plot predator (P)
    ax.plot(data['param_vals'], data['P_means'], '-', label='P: Stable', color='purple')
    ax.plot(data['param_vals'], data['P_cycle_center'], '--', label='P: Cycle Mean', color='purple')
    ax.fill_between(data['param_vals'], data['P_mins'], data['P_maxs'],
                    where=~np.isnan(data['P_cycle_center']), color='gray', alpha=0.3)

    ax.set_ylim(0, 9000)
    ax.set_ylabel("Abundance", fontsize=11)

    # Replace _H with _N
    display_name = param_name.replace("_H", "_N")
    ax.set_xlabel(display_name, fontsize=11)
    ax.set_title(f'Final Abundance Across {display_name}', fontsize=13)
    ax.legend(fontsize=8)

# Hide any unused subplot axes
for ax in axs[len(param_sweeps):]:
    ax.axis('off')

plt.tight_layout()
#plt.savefig("manuscript_figs/bifurcation_3x2_grid.jpg", dpi=300, bbox_inches='tight')
plt.show()
```
