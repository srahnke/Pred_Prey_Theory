```python
def dNdt(H, P, parameters_dict):
    #Unpacking herbivore parameters:
    r_H = parameters_dict['r_H'] # growth rate of herbivores
    K_H = parameters_dict['K_H'] # carrying capacity of herbivores
    theta_H = parameters_dict['theta_H'] # immigration of herbivores
    c = parameters_dict['c'] # addition to predator from prey
    d = parameters_dict['d'] # value of H when predation half its max value
    q_H = parameters_dict['q_H'] # proportionality constant for herbivore harvest
    E_H = parameters_dict['E_H'] # Effort for herbivore harvest
    #Unpacking predator parameters:
    r_P = parameters_dict['r_P'] # growth rate of predator 
    K_P = parameters_dict['K_P'] # carrying capactiy of predators 
    b = parameters_dict['b'] # mortality to prey from predator
    theta_P = parameters_dict['theta_P'] # immigration of predators  
    q_P = parameters_dict['q_P'] # proportionality constant for predators harvest
    E_P = parameters_dict['E_P'] # Effort for predator harvest
    
    #Calculate the derivative 
    dH = r_H*H*(1-H/K_H) - (c*H*P)/(d+H) + theta_H - q_H*E_H*H
    dP = r_P*P*(1-P/K_P) + (b*H*P)/(d+H) + theta_P - q_P*E_P*P
    
    return dH, dP
```

```python
# Function: numerical solver, 2nd order RK 

def RK2(H, P, parameters_dict):
    H_init = H
    P_init = P 
    
    dH1, dP1 = dNdt(H,P,parameters_dict)
    
    H1 = H + 0.5*dH1
    P1 = P + 0.5*dP1
    
    dH2, dP2 = dNdt(H1,P1,parameters_dict)
    
    dH_ave = (dH1+dH2)/2 
    dP_ave = (dP1+dP2)/2 
    
    H = H_init + dH_ave
    P = P_init + dP_ave
    
    #Calculate harvest 
    q_H = parameters_dict['q_H'] #proportionality constant for herbivore harvest 
    E_H = parameters_dict['E_H'] #Effort for herbivore harvest
    
    q_P = parameters_dict['q_P'] #proportionality constant for herbivore harvest 
    E_P = parameters_dict['E_P'] #Effort for herbivore harvest
    
    herbivore_harvest = q_H*E_H*H
    predator_harvest = q_P*E_P*P
    
    return H, P, herbivore_harvest, predator_harvest
```

```python
# Function to run the model 

def run_model(INIT_H, INIT_P, parameters_dict):
    
    numyears = parameters_dict['numyears']
    
    H_array = np.zeros(numyears+1)
    P_array = np.zeros(numyears+1)
    H_harvest_array = np.zeros(numyears+1)
    P_harvest_array = np.zeros(numyears+1)
    
    H_array[0] = INIT_H #Starting biomass of herbivore 
    P_array[0] = INIT_P #Starting biomass of predator 
    H_harvest_array[0] = 0
    P_harvest_array[0] = 0
    
    for i in np.arange(0,numyears):
        H_array[i+1], P_array[i+1], H_harvest_array[i+1], P_harvest_array[i+1] \
            = RK2(H_array[i], P_array[i], parameters_dict)
        i = i+1
    
    return H_array, P_array, H_harvest_array, P_harvest_array
```

```python
def default_params():
    numyears = 100
    #Herbivore parameters:
    r_H = 0.4 # growth rate of herbivores
    K_H = 5000 # carrying capacity of herbivores
    theta_H = 50 # immigration of herbivores
    c = 0.1 #additional mortality of H 
    d = 500 # value of H when predation half its max value 
    q_H = 1. # proportionality constant for herbivore harvest 
    E_H = 0.0 # Effort for herbivore harvest (0.05 = limit cycle, 0.2 = stable state)
    #Predator parameters:
    r_P = 0.2 # growth rate of predator 
    K_P = 2000 # carrying capactiy of predators
    b = 0.1 # additional growth of P due to H
    theta_P = 10 # immigration of predators 
    q_P = 1. # proportionality constant for predator harvest 
    E_P = 0.0 # Effort for predator harvest

    parameters_dict = {'numyears': numyears, 
                  'r_H': r_H, 
                  'K_H': K_H, 
                  'theta_H': theta_H,
                  'c': c, 
                  'd': d, 
                  'E_H': E_H,
                  'q_H': q_H, 
                  'r_P': r_P,
                  'K_P': K_P,
                  'theta_P': theta_P, 
                  'b': b,
                  'q_P': q_P, 
                  'E_P': E_P,
                  }

    return parameters_dict
```

```python
#Run model simulation; set params to default
parameters_dict = default_params()

numyears = 100
H0 = 100
P0 = 10

H_array, P_array, H_harvest_array, P_harvest_array = run_model(H0, P0, parameters_dict)
```

```python
# Define the predator-prey model
def dNdt(state, t, parameters_dict):
    H, P = state
    r_H = parameters_dict['r_H']
    K_H = parameters_dict['K_H']
    theta_H = parameters_dict['theta_H']
    c = parameters_dict['c']
    d = parameters_dict['d']
    q_H = parameters_dict['q_H']
    E_H = parameters_dict['E_H']
    r_P = parameters_dict['r_P']
    K_P = parameters_dict['K_P']
    b = parameters_dict['b']
    theta_P = parameters_dict['theta_P']
    q_P = parameters_dict['q_P']
    E_P = parameters_dict['E_P']
    
    dH = r_H * H * (1 - H / K_H) - (c * H * P) / (d + H) + theta_H - q_H * E_H * H
    dP = r_P * P * (1 - P / K_P) + (b * H * P) / (d + H) + theta_P - q_P * E_P * P
    return [dH, dP]

# Define prey nullcline (dH/dt = 0)
def prey_nullcline(H, parameters_dict):
    r_H = parameters_dict["r_H"]
    K_H = parameters_dict["K_H"]
    theta_H = parameters_dict["theta_H"]
    c = parameters_dict["c"]
    d = parameters_dict["d"]
    q_H = parameters_dict["q_H"]
    E_H = parameters_dict["E_H"]

    # Avoid division by zero
    H_safe = np.maximum(H, 1e-10)

    # Solve for P
    P_positive = (r_H * H_safe * (1 - H_safe / K_H) + theta_H - q_H * E_H * H_safe) * (d + H_safe) / (c * H_safe)
    return P_positive

# Define predator nullcline (dP/dt = 0)
def predator_nullcline(H, parameters_dict):
    r_P = parameters_dict["r_P"]
    K_P = parameters_dict["K_P"]
    b = parameters_dict["b"]
    d = parameters_dict["d"]
    theta_P = parameters_dict["theta_P"]
    q_P = parameters_dict["q_P"]
    E_P = parameters_dict["E_P"]

    # Avoid division by zero for H = 0
    H_safe = np.maximum(H, 1e-10)

    # Solve for P iteratively
    P_nullcline = []
    for H_i in H_safe:
        def predator_nullcline_func(P):
            return (
                r_P * P * (1 - P / K_P)
                + b * H_i * P / (d + H_i)
                + theta_P
                - q_P * E_P * P
            )
        # Solve using numerical methods
        P_solution = optimize.fsolve(predator_nullcline_func, [0.1, 10])  # Two roots
        P_nullcline.append(P_solution)

    return np.array(P_nullcline).T  # Return two branches

# Define parameters
parameters_dict = default_params()

# Create grid for phase plane analysis
H = np.linspace(1, 8000, 50)  # Herbivore population
P = np.linspace(1, 8000, 50)  # Predator population
H_grid, P_grid = np.meshgrid(H, P)

# Compute vector field
dH_grid, dP_grid = np.zeros(H_grid.shape), np.zeros(P_grid.shape)
for i in range(H_grid.shape[0]):
    for j in range(H_grid.shape[1]):
        dH_grid[i, j], dP_grid[i, j] = dNdt([H_grid[i, j], P_grid[i, j]], 0, parameters_dict)

# Normalize vectors for plotting
magnitude = np.sqrt(dH_grid**2 + dP_grid**2)
dH_grid /= magnitude
dP_grid /= magnitude

# Compute nullclines
H_range = np.linspace(1, 8000, 500)
P_prey_nullcline = prey_nullcline(H_range, parameters_dict)
P_predator_nullcline = predator_nullcline(H_range, parameters_dict)

# Plot vector field
plt.figure(figsize=(10, 7))
plt.quiver(H_grid, P_grid, dH_grid, dP_grid, color="gray", alpha=0.6, angles="xy")
plt.xlabel("Herbivore Population (H)")
plt.ylabel("Predator Population (P)")
plt.title("Phase Plane with Nullclines")

# Plot nullclines
plt.plot(H_range, P_prey_nullcline, 'g-', label="Prey Nullcline (dH/dt = 0)")
plt.plot(H_range, P_predator_nullcline[0], 'm--', label="Predator Nullcline (Lower Branch, dP/dt = 0)")
plt.plot(H_range, P_predator_nullcline[1], 'm-', label="Predator Nullcline (Upper Branch, dP/dt = 0)")

# Solve and overlay trajectories for different initial conditions
initial_conditions = [(100, 10), (1000, 200), (4000, 500), (500, 50)]
t = np.linspace(0, 200, 1000)  # Time array

for state0 in initial_conditions:
    trajectory = odeint(dNdt, state0, t, args=(parameters_dict,))
    plt.plot(trajectory[:, 0], trajectory[:, 1], label=f"IC: {state0}")

# Add legend and grid
plt.legend()
plt.xlim([1, 8000])
plt.ylim([1, 8000])
plt.show()
```

```python
# Define model parameters
def default_params():
    return {
        'r_H': 0.1,   # Prey growth rate
        'K_H': 1000,  # Prey carrying capacity
        'c': 0.1,     # Prey-predator interaction rate
        'd': 10,      # Prey handling time
        'theta_H': 0.1,  # Prey constant mortality rate
        'q_H': 0.05,    # Prey harvest efficiency
        'E_H': 0.1,     # Prey harvest intensity
        'r_P': 0.05,    # Predator growth rate
        'K_P': 1000,    # Predator carrying capacity
        'b': 0.1,       # Predator-prey interaction rate
        'theta_P': 0.1, # Predator constant mortality rate
        'q_P': 0.05,    # Predator harvest efficiency
        'E_P': 0.1      # Predator harvest intensity
    }

# Define the prey nullcline equation (dH/dt = 0)
def prey_nullcline(H, P, parameters_dict):
    r_H = parameters_dict['r_H']
    K_H = parameters_dict['K_H']
    c = parameters_dict['c']
    d = parameters_dict['d']
    theta_H = parameters_dict['theta_H']
    q_H = parameters_dict['q_H']
    E_H = parameters_dict['E_H']

    # Prey nullcline equation
    return r_H * H * (1 - H / K_H) - (c * H * P) / (d + H) + theta_H - q_H * E_H * H

# Define the predator nullcline equation (dP/dt = 0)
def predator_nullcline(H, P, parameters_dict):
    r_P = parameters_dict['r_P']
    K_P = parameters_dict['K_P']
    b = parameters_dict['b']
    d = parameters_dict['d']
    theta_P = parameters_dict['theta_P']
    q_P = parameters_dict['q_P']
    E_P = parameters_dict['E_P']

    # Predator nullcline equation
    return r_P * P * (1 - P / K_P) + (b * H * P) / (d + H) + theta_P - q_P * E_P * P

# Solve for the prey nullcline as a function of H and P
def solve_prey_nullcline(H_range, parameters_dict):
    P_nullcline = []
    for H in H_range:
        # Solve for P numerically (Prey nullcline equation)
        P_solution = optimize.fsolve(prey_nullcline, 10, args=(H, parameters_dict))  # Initial guess of 10
        P_nullcline.append(P_solution[0])
    return np.array(P_nullcline)

# Solve for the predator nullcline as a function of H and P
def solve_predator_nullcline(H_range, parameters_dict):
    P_nullcline = []
    for H in H_range:
        # Solve for P numerically (Predator nullcline equation)
        P_solution = optimize.fsolve(predator_nullcline, 10, args=(H, parameters_dict))  # Initial guess of 10
        P_nullcline.append(P_solution[0])
    return np.array(P_nullcline)

# Define parameter set
parameters_dict = default_params()

# Create H-range for nullcline calculation
H_range = np.linspace(1, 1000, 500)  # Avoid H=0 since the nullcline is trivial there

# Solve for nullclines
P_prey_nullcline = solve_prey_nullcline(H_range, parameters_dict)
P_predator_nullcline = solve_predator_nullcline(H_range, parameters_dict)

# Plotting
plt.figure(figsize=(10, 7))

# Plot the prey and predator nullclines
plt.plot(H_range, P_prey_nullcline, 'g-', label="Prey Nullcline (dH/dt = 0)")
plt.plot(H_range, P_predator_nullcline, 'm-', label="Predator Nullcline (dP/dt = 0)")

# Customize the plot
plt.xlabel("Herbivore Population (H)")
plt.ylabel("Predator Population (P)")
plt.title("Predator-Prey Model Nullclines")
plt.legend()
plt.grid(True)
plt.xlim([1, 1000])
plt.ylim([1, 1000])

plt.show()
```

```python
# Defining the system of equations
def model(t, z, rh, kh, c, d, rp, kp, b, qh, Eh, qp, Ep):
    H, P = z
    dHdt = (rh * H * ((kh - H) / kh)) - ((c * H * P) / (d + H)) - (qh * Eh * H)
    dPdt = (rp * P * ((kp - P) / kp)) + ((b * H * P) / (d + H)) - (qp * Ep * P)
    return [dHdt, dPdt]

# Function to solve the system
def solve_system(b, c, rh=0.4, kh=5000, d=500, rp=0.2, kp=2000, qh=1.0, Eh=0.0, qp=1.0, Ep=0.0, h0=100, p0=10, tmax=500):
    # Solve the system using scipy.integrate.solve_ivp
    t_span = (0, tmax)
    y0 = [h0, p0]
    solution = solve_ivp(model, t_span, y0, args=(rh, kh, c, d, rp, kp, b, qh, Eh, qp, Ep), max_step=0.0001)
    return solution

# Parameter combinations
parameter_combinations = [(0.1, 0.1), (0.3, 0.3), (0.6, 0.6)]

# Generate and plot solutions for each parameter combination
fig, axes = plt.subplots(len(parameter_combinations), 1, figsize=(10, 6), sharex=True, sharey=True)

for i, (b_value, c_value) in enumerate(parameter_combinations):
    solution = solve_system(b_value, c_value)
    
    # Extracting time and solutions
    t = solution.t
    H = solution.y[0]
    P = solution.y[1]
    
    # Plotting results
    axes[i].plot(t, H, label="Prey", color="blue")
    axes[i].plot(t, P, label="Predator", color="red")
    axes[i].set_title(f"b = {b_value}, c = {c_value}")
    axes[i].set_xlabel("Time in Years")
    axes[i].set_ylabel("Abundance")
    axes[i].legend()

plt.tight_layout()
plt.show()

# Nullclines and vector field plotting
def plot_nullclines_and_vector_field(rh, kh, c, d, rp, kp, b, qh, Eh, qp, Ep):
    x_vals = np.linspace(0, 8000, 100)
    y_vals = np.linspace(0, 8000, 100)
    X, Y = np.meshgrid(x_vals, y_vals)

    # Nullcline for Prey (H)
    dHdt = (rh * X * ((kh - X) / kh)) - ((c * X * Y) / (d + X)) - (qh * Eh * X)
    # Nullcline for Predator (P)
    dPdt = (rp * Y * ((kp - Y) / kp)) + ((b * X * Y) / (d + X)) - (qp * Ep * Y)

    fig, ax = plt.subplots(figsize=(6, 6))
    ax.contour(X, Y, dHdt, levels=[0], colors="blue", label="Prey Nullcline")
    ax.contour(X, Y, dPdt, levels=[0], colors="red", label="Predator Nullcline")

    # Adding vector field
    DX = (rh * X * ((kh - X) / kh)) - ((c * X * Y) / (d + X)) - (qh * Eh * X)
    DY = (rp * Y * ((kp - Y) / kp)) + ((b * X * Y) / (d + X)) - (qp * Ep * Y)
    ax.quiver(X, Y, DX, DY, color="gray", headlength=3, scale=30)

    ax.set_xlabel("Prey Abundance")
    ax.set_ylabel("Predator Abundance")
    ax.set_title(f"Nullclines and Vector Field: b = {b}, c = {c}")
    ax.legend()
    plt.show()

# Example for a parameter combination
plot_nullclines_and_vector_field(0.1, 0.1, 0.1, 500, 0.2, 2000, 0.1, 1.0, 0.0, 1.0, 0.0)
```
